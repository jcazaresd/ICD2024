# Análisis exploratorio de Dataset

Esta practica consiste en seleccionar un Dataset de un tema de interés que incluya diferentes variables que puedan ser analizadas por medio las librerías de python. La base de datos seleccionada para el desarrollo del análisis exploratorio se titula **World's Air Quality and Water Pollution Dataset** (Calidad del aire y contaminación del agua en el mundo) y fue obtenida de la plataforma [Kaggle](https://www.kaggle.com/datasets/victorahaji/worlds-air-quality-and-water-pollution-dataset). En cuanto a la calidad del aire, los valores están representados en una escala del 0 al 100, donde 0 indica la peor calidad posible y 100 representa la mejor calida. dCaso contrario para los valores de contaminación del agua, en los que 0 indica un valor mínimo al.

# ANÁLISIS EXPLORATORIO 
#### PASO 4
Como se ha establecido, la cantidad de datos presentada es sumamente grande, dificultando el análisis. Debido a esto se aplicará un **filtro** para las ciudades que se encuentran en México. 

```python
filtered_df = df[df[' "Country"'] == ' "Mexico"']
print (filtered_df)
```

#### PASO 5

De manera inicial para el análisis exploratorio de los datos filtrados, se graficará un diagrama de barras, los datos de calidad del aire y a su vez del agua. Esto para poder determinar de forma general cuales son las ciudades que presentan los valores críticos de ambas variables

```python
sns.barplot(y='City', x=' "AirQuality"', data=filtered_df)
plt.title("Calidad del aire por ciudad")
plt.show()

sns.barplot(y='City', x=' "WaterPollution"', data=filtered_df)
plt.title("Contaminación del agua por ciudad")
plt.show()
```

#### PASO 6 

Se procedió a realizar un diagrama de cajas y bigotes y un gráfico de dispersión con líneas de regresión al igual que con los datos totales para poder visualizar los cambios en las variables y establecer relaciones. 

```python
sns.boxplot(data=filtered_df)
```

```python
sns.lmplot(x=' "AirQuality"', y=' "WaterPollution"', data=filtered_df)
```

#### PASO 7 

Adicionalmente, y al contar con menos valores, se realizó un pairplot para visualizar las relaciones entre múltiples variables en un dataset mediante gráficos de dispersión y distribuciones univariadas.

```python
sns.pairplot(filtered_df, hue=' "Region"')
plt.show()
```

### CONCLUSIÓN 

las condiciones de **contaminación del agua** se disttibuyen entre valores que indican un grado de contaminación moderado (40) hasta elevado (70) En cuanto al nivel de calidad del aire, estos niveles varían desde la calidad moderada (50) hasta una calidad que se puede considerar como buena (80) 

Se observa una **relación positiva**, a medida que la calidad del aire mejora, también aumenta la contaminación del agua.
Los datos muestran una amplia variabilidad en ambos ejes, de 0 a 100. La línea de mejor ajuste refuerza la **correlación positiva** entre las dos variables. La zona sombreada alrededor de la línea de regresión indica la incertidumbre en la predicción.
La relación podría sugerir que las mejoras en la calidad del aire y la contaminación del agua están influenciadas por factores comunes, como políticas ambientales o actividades industriales.

# ANALISIS EXPLORATORIO 
#### PASO 8
Con el fin de poder determinar las ciudades con las condiciones de calidad de aire y del agua más deficientes, se aplicará un **segundo filtro** en el que se discriminen valores por debajo del 50 % para calidad del aire, y por encima del 50 % para contaminación del agua.

```python
filtered2 = filtered_df[(filtered_df[' "AirQuality"'] < 50) & (filtered_df[' "WaterPollution"'] > 50)]
print (filtered2)
```

#### PASO 9 

Se procedió a graficar los datos para poder identificar las ciudades en México en las que se presentan los valores más deficientes de calidad del aire y más grandes de contaminación del agua

```python
sns.barplot(y='City', x=' "AirQuality"', data=filtered2)
plt.title("Calidad del aire por ciudad")
plt.xlabel('Peor grado calidad del aire por ciudad')
plt.ylabel('Ciudad')
```

```python
sns.barplot(y='City', x=' "WaterPollution"', data=filtered2)
plt.title("Contaminación del agua por ciudad")
plt.xlabel('Mayor grado contaminación del agua por ciudad')
plt.ylabel('Ciudad')
```

```python
df_long = pd.melt(filtered2, value_vars=[' "AirQuality"', ' "WaterPollution"'], 
                  var_name='Tipo', value_name='Valor')

# Creamos el gráfico de violín
ax = sns.violinplot(data=df_long, x='Tipo', y='Valor', hue='Tipo', palette='muted')

plt.xlabel('Tipo')
plt.ylabel('Valor')
plt.title('Comparación de Calidad y Contaminación')
plt.show()
```

```python
sns.jointplot(data = filtered2,
              x = ' "AirQuality"',
              y = ' "WaterPollution"',
              kind = "reg")
```

#### PASO 10 

Una vez identificadas las ciudades críticas para el análisis, se procede a ordenar las ciudades, en forma ascendente para Calidad del aire y descendente para contaminación del agua. 

```python
sorted_df = filtered2.sort_values(by=' "AirQuality"', ascending=True) 
print (sorted_df)
```

```python
sorted_df = filtered2.sort_values(by=' "WaterPollution"', ascending=False) 
print (sorted_df)
```

## CONCLUSIÓN 

Aplicar el filtro de valores críticos y ordenarlos de tal manera en que se pueda establecer cuales son las ciudades con mayor contaminación del agua y menor calidad del aire, se estableció que para calidad del aire, **Irapuato** es la más crítica, seguida por Ciudad de México y Tijuana. En cuanto a la contaminación del agua se determinó que **Acapulco** es la ciudad con mayor contaminación, seguida por Ciudad de México y Tijuana. 

Conocer estos valores e identificar las ciudades puede promover la realización de acciones de remediación con especial atención a los sitios identificados.
